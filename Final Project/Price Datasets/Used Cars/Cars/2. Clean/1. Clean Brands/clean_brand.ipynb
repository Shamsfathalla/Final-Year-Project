```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import ast
import numpy as np

df = pd.read_csv('merged_all_used_cars.csv')
```

```python
print("--- ALL BRANDS ---")
brands = df['Brand'].dropna().unique()
brands.sort()
print(brands)
print(f"\nTotal Brands: {len(brands)}\n")
```

```python
# 1. Standardize casing to Title Case first (e.g., 'alfa romeo' -> 'Alfa Romeo')
# This fixes the majority of simple casing issues immediately.
df['Brand'] = df['Brand'].astype(str).str.strip().str.title()

# 2. Define a dictionary for specific corrections
# (Acronyms, merged names, spelling fixes, and removing hyphens)
brand_corrections = {
    # Acronyms & Uppercase
    'Bmw': 'BMW',
    'Byd': 'BYD',
    'Baic': 'BAIC',
    'Dfsk': 'DFSK',
    'Ds': 'DS',
    'Gac': 'GAC',
    'Gmc': 'GMC',
    'Jac': 'JAC',
    'Jmc': 'JMC',
    'Mg': 'MG',
    'Vgv': 'VGV',
    'Xpeng': 'XPeng',
    'Exeed': 'EXEED',
    
    # Merging Variations & Spelling Fixes
    'Alfa': 'Alfa Romeo',
    'Alfa-Romeo': 'Alfa Romeo',
    'Aston-Martin': 'Aston Martin',
    'Citroën': 'Citroen',
    'Great': 'Great Wall',
    'Great-Wall': 'Great Wall',
    'Land': 'Land Rover',
    'Land-Rover': 'Land Rover',
    'Mercedes': 'Mercedes-Benz', 
    'Mercedes-Benz': 'Mercedes-Benz', # Ensures consistency
    'Mini': 'Mini', # 'MINI' maps here via title()
    'Range': 'Land Rover', # Range Rover is a model of Land Rover
    'Range Rover': 'Land Rover',
    'Rolls': 'Rolls-Royce',
    'Rolls Royce': 'Rolls-Royce',
    'Ssang-Yong': 'SsangYong',
    'Ssang Yong': 'SsangYong',
    'Ssangyong': 'SsangYong'
}

# 3. Apply the corrections
df['Brand'] = df['Brand'].replace(brand_corrections)

# 4. Check the results
print("--- ALL BRANDS ---")
brands = df['Brand'].dropna().unique()
brands.sort()
print(brands)
print(f"\nTotal Brands: {len(brands)}\n")
```

```python
df.to_csv('merged_all_used_cars_cleaned1.csv', index=False)
print("✓ Data saved to 'merged_all_used_cars_cleaned1.csv'")
```
