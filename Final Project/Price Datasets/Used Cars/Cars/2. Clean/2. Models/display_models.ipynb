```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import ast
import os
import numpy as np

df = pd.read_csv('merged_all_used_cars_cleaned1.csv', encoding='latin-1')
```

```python
# Replace actual NaNs with a label so they aren't skipped in the loop
display_df = df[['Brand', 'Model']].fillna('(Empty)')

# Drop duplicates so we only see unique combinations
grouped = display_df.drop_duplicates().sort_values(['Brand', 'Model'])

print("\n--- VEHICLE HIERARCHY (Brand and Model Only) ---")
for brand, brand_df in grouped.groupby('Brand'):
    print(f"\nBRAND: {brand}")
    for model in brand_df['Model'].tolist():
        print(f"  └─ {model}")
```

```python
trim_counts = (
    df["Trim"]
    .fillna("Unspecified")
    .astype(str)
    .str.strip()
    .value_counts()
)

print(trim_counts)
```

```python
df["Brand"] = df["Brand"].astype(str).str.strip()

BASE_DIR = "cars_by_brand"
os.makedirs(BASE_DIR, exist_ok=True)

def clean_name(name):
    name = re.sub(r"[\\/*?:\"<>|]", "", name)
    return name.strip()

for brand, brand_df in df.groupby("Brand"):
    safe_brand = clean_name(brand)

    brand_dir = os.path.join(BASE_DIR, safe_brand)
    os.makedirs(brand_dir, exist_ok=True)

    csv_path = os.path.join(brand_dir, f"{safe_brand}.csv")
    brand_df.to_csv(csv_path, index=False, encoding="utf-8-sig")
```
